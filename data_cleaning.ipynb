```python
# imports library 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
file_path ='../Carbon_emission_project/Data_files/energy_consumption.csv'
df_1= pd.read_csv(file_path, skiprows=2, encoding='unicode_escape')
df_1.head()
```

```python
df_1.dropna(inplace=True)
```

```python
df_1.rename(columns={'Exajoules':'Country'}, inplace=True)
df_1.head(15)
```

```python
df_1 = df_1[~df_1['Country'].str.contains('Total|total', case=False, na=False)]
```

```python
df_1.head(30)
```

```python
df_1 = df_1[~df_1.apply(lambda row: row.astype(str).str.contains('-').any(), axis=1)]
```

```python
#df_1 = df_1[~df_1.apply(lambda row: row.astype(str).str.contains('^').any(), axis=1)]
```

```python
df_1.replace('^', 0, inplace=True)
```

```python
df_1.replace('?', 0, inplace=True)
```

```python
df_1.tail(30)
```

```python
last_two_indices = df_1.index[-2:]
df_1.drop(last_two_indices, inplace=True)
```

```python
df_1.drop(index=1, inplace=True)
```

```python
df_1.to_csv('../Carbon_emission_project/Data_files/clean_energy_consumption.csv', index=False)
```
